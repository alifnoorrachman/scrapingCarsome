```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Function to extract Product Title
def get_title(soup):

    try:
        # Outer Tag Object
        title = soup.find("a", attrs={"class":"mod-b-card__title"})

        # Inner NavigatableString Object
        title_value = title.text

        # Title as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string
```

```python
# Function to extract Product Title
def get_price(soup):

    try:
        # Outer Tag Object
        title = soup.find("div", attrs={"class":"mod-card__price-cash"})

        # Inner NavigatableString Object
        title_value = title.text

        # Title as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string
```

```python
# Function to extract Product Title
def get_km(soup):

    try:
        # Outer Tag Object
        title = soup.find("div", attrs={"class":"mod-b-card__car-other"})

        # Inner NavigatableString Object
        title_value = title.text

        # Title as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string
```

```python
if __name__ == '__main__':

    # add your user agent
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36','Accept-Language': 'en-US, en;q=0.5', 'Accept-Language': 'en-US, en;q=0.5'})

    # The webpage URL
    URL = "https://www.carsome.id/en/buy-car?bodyType=2"

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch links as List of Tag Objects
    links = soup.find_all("div", attrs={"class":"list-card__item"})

    # Store the links
    links_list = []

    # Loop for extracting links from Tag Objects
    for link in links:
            links_list.append(link.get('href'))

    d = {"title":[],"km":[],"price":[]}

    # Loop for extracting product details from each link
    for link in links_list:
        new_webpage = requests.get("https://www.carsome.id", headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['title'].append(get_title(new_soup))
        d['km'].append(get_km(new_soup))
        d['price'].append(get_price(new_soup))


df = pd.DataFrame.from_dict(d)
df['title'].replace('', np.nan, inplace=True)
df = df.dropna(subset=['title'])
```
